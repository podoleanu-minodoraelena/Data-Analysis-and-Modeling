```python
X = df.drop('Placement_Yes',axis=1) 
y = df['Placement_Yes']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=21)
```

```python
# --------------------------
# Scalarea dataset-ului folosind MinMaxScaler
# --------------------------

from sklearn.preprocessing import MinMaxScaler

# Crearea obiectului scaler
scaler = MinMaxScaler()

# Învățarea min și max pe baza întregului set X
scaler.fit(X)

# Aplicare scaler pe setul de antrenament
X_train = scaler.fit_transform(X_train)

# Aplicare scaler pe setul de test
X_test = scaler.transform(X_test)

# Transformarea rezultatul în DataFrame pentru lizibilitate, păstrând numele coloanelor
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)

# Afișare primelor 5 rânduri din seturile scalate
print('========== Antrenare ==========\n')
X_train.head()
```

```python
print('========== Testare ==========\n')
X_test.head()
```

```python
X_test.size
```

```python
X_train.size
```

Doar 3 coloane sunt categoriale, restul numerice. Dimensiunea seturilor de antrenare 14000, respectiv de testare 56000.

```python
# -----------------------------
# Gaussian Naive Bayes
# -----------------------------

from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Crearea modelului Gaussian Naive Bayes
gaussian = GaussianNB()

# Antrenarea modelului pe datele de training
gaussian.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_NB = gaussian.predict(X_test)

# Calcularea acurateții pe setul de training
gaussian_train = round(gaussian.score(X_train, y_train) * 100, 2)

# Calcularea acurateții pe setul de test
gaussian_accuracy = round(accuracy_score(y_pred_NB, y_test) * 100, 2)

# Calcularea scorului F1 pe setul de test
gaussian_f1 = round(f1_score(y_pred_NB, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", gaussian_train, "%")
print("Acuratețea modelului           :", gaussian_accuracy, "%")
print("Scorul F1                      :", gaussian_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_NB))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score

# -----------------------------
# Matricea de confuzie pentru modelul de regresie logistică
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 


# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_NB)

# Transformarea matricei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matricea de Confuzie - Gaussian Naive Bayes")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

```python
# -----------------------------
# Curba ROC - Gaussian Naive Bayes
# -----------------------------
from sklearn.metrics import RocCurveDisplay
plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza estimatorului (modelului)
RocCurveDisplay.from_estimator(
    gaussian,  # modelul Gaussian Naive Bayes
    X_test,    # datele de test
    y_test     # valorile reale
)

# Titlul graficului
plt.title("Curba ROC - Gaussian Naive Bayes")

# Afișare figură
plt.show()
```

```python
# -----------------------------
# Curba Precizie-Recall - Gaussian Naive Bayes
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei PR pe baza estimatorului (modelului)
PrecisionRecallDisplay.from_estimator(
    gaussian,   # modelul Gaussian Naive Bayes
    X_test,     # datele de test
    y_test      # valorile reale
)

# Titlul graficului 
plt.title("Curba Precizie-Recall - Gaussian Naive Bayes")

# Afișare figură
plt.show()
```

Acuratețea pe setul de training este 89.86 %
Scorul F1 este 55.98 %.
Matricea de confuzie arată că modelul Naive Bayes clasifică corect majoritatea observațiilor, cu o acuratețe totală de aproximativ 89%. Se observă un dezechilibru între erorile de tip fals pozitiv (19) și fals negativ (198).  Curba ROC pentru modelul Naive Bayes indică o performanță  bună, având o valoare AUC de 0.79. Acest rezultat arată că modelul are o capacitate ridicată de diferențiere între clasele pozitive și negative, confirmând calitatea generală bună a clasificatorului observată și în celelalte metrici.

```python
# -----------------------------
# Decision Tree Classifier
# -----------------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Crearea modelul Decision Tree
decision = DecisionTreeClassifier()

# Antrenarea modelului pe datele de training
decision.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_Decision = decision.predict(X_test)

# Calcularea acurateții pe setul de training
decision_train = round(decision.score(X_train, y_train) * 100, 2)

# Calcularea acuratețeii pe setul de test
decision_accuracy = round(accuracy_score(y_pred_Decision, y_test) * 100, 2)

# Calcularea scorului F1 pe setul de test
decision_f1 = round(f1_score(y_pred_Decision, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", decision_train, "%")
print("Acuratețea modelului           :", decision_accuracy, "%")
print("Scorul F1                      :", decision_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_Decision))
```

```python
# -----------------------------
# Matricea de confuzie - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_Decision)

# Transformarea matriceei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matricea de Confuzie - Decision Tree")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

```python
# -----------------------------
# Curba ROC - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza modelului Decision Tree
RocCurveDisplay.from_estimator(
    decision,       # modelul antrenat
    X_test,    # datele de test
    y_test     # valorile reale
)

plt.title("Curba ROC - Decision Tree")

# Afișare figură
plt.show()
```

```python
# -----------------------------
# Curba Precizie-Recall -  Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei Precizie-Recall
PrecisionRecallDisplay.from_estimator(
    decision,       # modelul antrenat
    X_test,    # datele de test
    y_test     # valorile reale
)

plt.title("Curba Precizie-Recall - Decision Tree")

#Afișare figură
plt.show()
```

Acuratețea pe setul de training este 89.86 % Scorul F1 este 55.98 % pentru Naive Bayes, iar pentru arborele de decizie ambele sunt de 100%. Analiza suplimentară, inclusiv matricea de confuzie, curbele ROC și PR, confirmă ca GaussianNB oferă o discriminare mai echilibrată între clase, în timp ce Decision Tree, deși performant pe training, este mai puțin robust pe date noi.
